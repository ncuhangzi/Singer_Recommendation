```python
import pandas as pd
import numpy as np
```

```python
#------------------
# LOAD THE DATASET
#------------------

data = pd.read_csv('lastfm.csv')

# Create a new dataframe without the user ids.
data_items = data.drop('user', 1)
```

```python
data_items.loc[1:, 'yann tiersen':]*0
data_items['yann tiersen'].apply(lambda x: 0 if x==0 else 0, inplace=True)
data_items.loc[1:, 'yann tiersen':].head(30)


   
```

```python
#把其中一欄歌手值拿掉的做法
data_test = data_items.loc[:, 'yann tiersen':]
data_r = data_items.loc[1:, 'yann tiersen':].mul(0)
data_items = data_items.drop('yann tiersen',axis=1)
#這邊不能用concat 他會把df上下接在一起
pd.concat([data_items, data_r], ignore_index=True)

```

```python
#用numpy吧...不會寫numpy啦啦啦
data_items.loc[:, '':] = np.()
```

```python
# As a first step we normalize the user vectors to unit vectors.

# magnitude = sqrt(x2 + y2 + z2 + ...)
magnitude = np.sqrt(np.square(data_items).sum(axis=1))

# unitvector = (x / magnitude, y / magnitude, z / magnitude, ...)
data_items = data_items.divide(magnitude, axis='index')
```

```python
data_items.tail(5)
```

```python
#------------------
#SVD Decomposition
#------------------
#用scipy的svd function 可以自行選擇設定潛在因素(lantent factor U, sigma, VT)
#R=UΣVT  保留前k個最重要的因素
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(data_items, k = 50)
# the Σ returned is just the values instead of a diagonal matrix
sigma = np.diag(sigma)
print(sigma)
```

```python
#------------------
#Making Predictions from Decomposed Matrices 
#------------------+ magnitude.reshape(-1, 1)
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) 
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = data_items.columns)
```
